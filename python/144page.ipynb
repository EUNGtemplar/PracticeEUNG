```python
url_hollys = "https://www.hollys.co.kr/store/korea/korStore2.do?pageNo=51&sido=&gugun=&store=#"
print(url_hollys)
```

```python

for page in range(1, 52): #50*10 + 4 = 504줄 
    url_hollys = f"https://www.hollys.co.kr/store/korea/korStore2.do?pageNo={page}&sido=&gugun=&store=#"
    print(url_hollys)
```

```python
shopinfo = []

for page in range(1, 52): #50*10 + 4 = 504줄 
    url_hollys = f"https://www.hollys.co.kr/store/korea/korStore2.do?pageNo={page}&sido=&gugun=&store=#"
    print(url_hollys)

    soup = BeautifulSoup(requests.get(url_hollys).text, "html.parser")
    tbody = soup.find("tbody")
    trs = tbody.find_all("tr")
    for tr in trs:
        tds = tr.find_all("td")
        shopinfo.append([tds[0].string,
                         tds[1].string,
                         tds[3].string,
                         tds[5].string])
shopinfo
```

```python
len(shopinfo)
```

```python
#!pip install pandas
import pandas as pd
```

```python
df_shopinfo = pd.DataFrame(shopinfo)
df_shopinfo
```

```python
df_shopinfo.to_csv("hollys_shopinfo.csv", encoding="cp949")
```
